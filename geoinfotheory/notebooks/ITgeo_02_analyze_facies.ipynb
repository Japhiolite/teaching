## Analyze results of generated facies fields

In this step, we are now using the generated facies fields (from previous notebook).


```python
# basic imports
import numpy as np                                        
import matplotlib.pyplot as plt     
import pickle
```

### Load Data

We first load the dataset from the generated `pickle` file:

```python
realizations = pickle.load(open("facies_realizations_ncut_64x64.pkl", 'rb'))
```

### Plot realizations

Just to get an impression, here a plot of some realizations:

```python
plt.figure(figsize=(12,6))
for i in range(1, 9):
    plt.subplot(2, 4, i)
    plt.imshow(realizations[i])
```

...and here simply the sum:

```python
plt.imshow(np.sum(realizations, axis=0))
plt.colorbar()
```

...and the (estimated) probability for facies=1:

```python
plt.imshow(np.sum(realizations, axis=0)/len(realizations))
plt.colorbar()
```
