__Python history__

First version was used in 1989  
The developer released the source code (version 0.9.0) in 1991  
Version 1.0 was completed in 1994  
Version 2.0 was released in the early 2000's  
Version 3.0 was released in 2008:  
> Complete overhaul which fixed many of the core languages flaws, in doing so, backwards compatbility was broken.  
> There are some compatbility issues you may come across from time to time, as many existing Python code is written in version 2x, but version 2x will eventually be phased out and future improvements and libraries will be written for Version 3x
